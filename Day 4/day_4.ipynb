# Problem 1

```python
res = []
with open('input_day_4.txt','r') as f:
    # Use list comprehension to read each character in file into 2D matrix, 
    # replacing tree by "1" and open field by "0"
    for i in f.readlines():
        res.append(i)
```

```python
res
```

```python
mandatory_fields = ['byr', 'iyr', 'eyr', 'hgt', 'hcl', 'ecl', 'pid']
```

```python
passports = []
passports_count = []
fields = []
for i in res:
    if i == '\n':
        tmp = 0
        for j in fields:
            if j.split(":")[0] in mandatory_fields:
                tmp += 1
        passports_count.append(tmp)
        passports.append(fields)
        fields = []
    else:
        fields+= i.split()
        
tmp = 0
for j in fields:
    if j.split(":")[0] in mandatory_fields:
        tmp += 1
passports_count.append(tmp)
passports.append(fields)
    
```

```python
passports
```

```python
passports_count = np.array(passports_count)
print((passports_count == 7).sum())
```

# Problem 2


```python
import re
```

```python
passports = np.array(passports)
passports = passports[passports_count == 7]
```

```python
passports.shape
```

```python
ecl = ['amb', 'blu', 'brn', 'gry', 'grn', 'hzl', 'oth']
passports_count = []
for passport in passports:
    cond = 0
    for fields in passport:
        field = fields.split(":")[0]
        data = fields.split(":")[1]
        cond += int(field == 'pid' and bool(re.match('^[0-9]{9}$',data)))\
            + int(field == 'ecl' and data in ecl)\
            + int(field == 'hcl' and bool(re.match('^\#[a-f0-9]{6}$',data)))\
            + int(field =='eyr' and bool(re.match('^[0-9]{4}$',data)) and 2020 <= int(data) <= 2030)\
            + int(field =='iyr' and bool(re.match('^[0-9]{4}$',data)) and 2010 <= int(data) <= 2020)\
            + int(field =='byr' and bool(re.match('^[0-9]{4}$',data)) and 1920 <= int(data) <= 2002)\
            + int(field =='hgt' and bool(re.match('^[0-9]{2}in$',data)) and 59 <= int(data[:-2]) <= 76)\
            + int(field =='hgt' and bool(re.match('^[0-9]{3}cm$',data)) and 150 <= int(data[:-2]) <= 193)
    passports_count.append(cond)
            
```

```python
passports_count = np.array(passports_count)
print((passports_count == 7).sum())
```
